## MOMA Art Data Set

I came across this data set on Twitter and thought that it would be interesting to explore it. [Here is the source](https://github.com/MuseumofModernArt/collection). As noted at the previous link, this data is not complete.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

There are a significant number of missing values. Let's soldier on. Title, MOMANumber, URL, CreditLine, and ObjectID don't seem to be very interesting columns by themselves. I'm going to work on a column by column basis from bottom to top.

How many of these artworks have been approved by curators?

```python
curator_approved = MOMA_artworks.CuratorApproved.value_counts()
ax = curator_approved.plot(kind='bar',
                           title='Curator Approved Artworks in MOMA Data Set')
ax.set_xticklabels(['No','Yes'],rotation=0)
```

So it looks fairly even - an assessment borne out by the math.

```python
100*MOMA_artworks.CuratorApproved.value_counts()/len(MOMA_artworks)
```
