## Exploring Climate Change Data

This data contains climate change data from May 1983 to December 2008, and comes from the ESRL/NOAA Physical Sciences Division. A description of the variables lies below.

>Year: the observation year.

>Month: the observation month.

>Temp: the difference in degrees Celsius between the average global temperature in that period and a reference value. This data comes from the Climatic Research Unit at the University of East Anglia.

>CO2, N2O, CH4, CFC.11, CFC.12: atmospheric concentrations of carbon dioxide (CO2), nitrous oxide (N2O), methane  (CH4), trichlorofluoromethane (CCl3F; commonly referred to as CFC-11) and dichlorodifluoromethane (CCl2F2; commonly referred to as CFC-12), respectively. This data comes from the ESRL/NOAA Global Monitoring Division.

>* CO2, N2O and CH4 are expressed in ppmv (parts per million by volume  -- i.e., 397 ppmv of CO2 means that CO2 constitutes 397 millionths of the total volume of the atmosphere)


>* CFC.11 and CFC.12 are expressed in ppbv (parts per billion by volume). 

>Aerosols: the mean stratospheric aerosol optical depth at 550 nm. This variable is linked to volcanoes, as volcanic eruptions result in new particles being added to the atmosphere, which affect how much of the sun's energy is reflected back into space. This data is from the Godard Institute for Space Studies at NASA.

>TSI: the total solar irradiance (TSI) in W/m2 (the rate at which the sun's energy is deposited per unit area). Due to sunspots and other solar phenomena, the amount of energy that is given off by the sun varies substantially with time. This data is from the SOLARIS-HEPPA project website.

>MEI: multivariate El Nino Southern Oscillation index (MEI), a measure of the strength of the El Nino/La Nina-Southern Oscillation (a weather effect in the Pacific Ocean that affects global temperatures).

```python
climate_data.Year.value_counts()
```

The first year is missing 4 records. Which months are they?

```python
climate_data[climate_data.Year==1983].Month.values
```

There's no data for 1983 from January to April. Perhaps the record keeping only started in May of that year. Now to the yearly data.

```python
climate_data_by_year = climate_data.groupby(['Year']).agg([np.median,max,min])
climate_data_by_year = climate_data_by_year.drop('Month',1)
climate_data_by_year.head()
```

```python
plt.plot(climate_data_by_year.index,climate_data_by_year.Temp['median'],label='median')
plt.plot(climate_data_by_year.index,climate_data_by_year.Temp['max'],label='max')
plt.plot(climate_data_by_year.index,climate_data_by_year.Temp['min'],label='min')
plt.xlim((1983,2008))
plt.legend(loc = 'upper left')
plt.xlabel('Year')
plt.title('Temperature Difference By Year')
plt.ylabel('Temperature Difference')
```

It seems that 1984 had the lowest temperature difference by some (relative) margin. The temperature difference fell below 0 for only a couple of years.

```python
climate_data_by_year.Temp.index[climate_data_by_year.Temp['min'] < 0]
```

The greatest temperature difference was recorded in what looks like 1998, and it's no surprise that the highest median difference is from that year as well.

```python
climate_data_by_year.Temp.index[climate_data_by_year.Temp['max']>.7]
```

How about the smallest and biggest fluctuation of temperature difference?

```python
temp_diff = []
for i in climate_data_by_year.index:
    temp_diff.append((i,climate_data_by_year.Temp['max'].ix[i] - climate_data_by_year.Temp['min'].ix[i]))
temp_diff.sort(key=lambda x: x[1])

print('The largest fluctuation was {0:.2f} in {1}.'.format(temp_diff[-1][1],temp_diff[-1][0]))
print('The smallest fluctuation was {0:.2f} in {1}.'.format(temp_diff[0][1],temp_diff[0][0]))
```

It might be worth it to take a glance at month by month trends just to see if anything interesting pops out.

```python
climate_data_by_month = climate_data.groupby(['Month']).agg(np.median)
climate_data_by_month = climate_data_by_month.drop('Year',1)
climate_data_by_month.head()
```

```python
fig = plt.figure()

for i, col in enumerate(climate_data_by_month.columns):
    plt.subplot('42'+str(i))
    plt.plot(climate_data_by_month.index,climate_data_by_month[col])
    plt.title(col)
    plt.gca().axes.get_yaxis().set_visible(False)
    
fig.subplots_adjust(hspace=1)
```

Trends are visible, but is the monthly data meaningful given that it's aggregated over more than 20 years? Hard to say.
